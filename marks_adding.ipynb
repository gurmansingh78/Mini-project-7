```python
import pandas as pd
df=pd.read_csv('/content/sample_data/Mini Project - Marks Adding.csv')
```

```python
df.head()
```

```python
df['Roll Num'].unique()
```

```python
roll_nos=list(df['Roll Num'].unique())
```

```python
output = {}

for r, m in zip(df['Roll Num'], df['Marks']):
    if r not in output:
        output[r] = 0
    output[r] += m

print(output)
```

```python
Output=pd.DataFrame(list(output.items()),columns=['Roll Num','Total Marks'])
```
